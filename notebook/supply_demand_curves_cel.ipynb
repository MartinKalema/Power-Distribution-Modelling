```python
import numpy as np
import pandas as pd
```

```python
min_allocated_power_cel = pd.read_excel('min_allocated_power_cel.xlsx')
max_allocated_power_cel = pd.read_excel('max_allocated_power_cel.xlsx')
max_hh_dd = pd.read_excel('max_hh_demand_reformatted.xlsx')
min_hh_dd = pd.read_excel('min_hh_demand_reformatted.xlsx')
demand_supply = pd.read_excel('Demand_Supply.xlsx')
```

```python
# sum columns for each table
sum_max_hh_dd = max_hh_dd.sum()
sum_min_hh_dd = min_hh_dd.sum()
sum_min_allocated_power_cel = min_allocated_power_cel.sum()
sum_max_allocated_power_cel = max_allocated_power_cel.sum()
```

```python
ss_power_values = demand_supply['SS_Power']
```

## Min Curve

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

# Plot the mean with a very thick line
ax.plot(range(1, 25), ss_power_values, color='red', linewidth=1, label='Total Power Supplied')
ax.plot(range(1, 25), sum_min_allocated_power_cel, color='green', linewidth=1, label='Power Allocated By CEL')
ax.plot(range(1, 25), sum_min_hh_dd, color='blue', linewidth=1, label='Minimum Hourly Demand', linestyle="--")

# Shade the space between the two lines
ax.fill_between(range(1, 25), ss_power_values, sum_min_allocated_power_cel, color='lightgray', alpha=0.3)

# Set the axis labels and title
ax.set_xlabel('Hour')
ax.set_ylabel('Power Supply')
ax.set_title('Power Supply against Power Allocated By CEL & Minimum Required')

# legend
ax.legend()

# save
plt.savefig('power_supply_against_power_allocated_by_cel_and_minimum_required.png')

# Show the plot
plt.show()
```

## Max Curve

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

# Plot the mean with a very thick line
ax.plot(range(1, 25), ss_power_values, color='red', linewidth=1, label='Total Power Supplied')
ax.plot(range(1, 25), sum_max_allocated_power_cel, color='green', linewidth=1, label='Power Allocated By CEL')
ax.plot(range(1, 25), sum_max_hh_dd, color='blue', linewidth=1, label='Maximum Hourly Demand', linestyle="--")

# Shade the space between the two lines
ax.fill_between(range(1, 25), ss_power_values, sum_max_allocated_power_cel, color='lightgray', alpha=0.3)

# Set the axis labels and title
ax.set_xlabel('Hour')
ax.set_ylabel('Power Supply')
ax.set_title('Power Supply against Power Allocated By CEL & Maximum Required')

# legend
ax.legend()

# save
plt.savefig('power_supply_against_power_allocated_by_cel_and_maximum_required.png')

# Show the plot
plt.show()
```
