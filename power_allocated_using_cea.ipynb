```python
import pandas as pd
import numpy as np
```

```python
from IPython.display import display

display.max_columns = None  
display.max_rows = None   

np.set_printoptions(threshold=np.inf)
```

```python
demand_supply = pd.read_excel('Demand_Supply.xlsx')
avg_hh_dd = pd.read_excel('avg_hh_demand_reformatted.xlsx')
max_hh_dd = pd.read_excel('max_hh_demand_reformatted.xlsx')
min_hh_dd = pd.read_excel('min_hh_demand_reformatted.xlsx')
```

## Dividing SS_Power by 156 (number of households)

```python
ss_power = demand_supply['SS_Power'].values
```

```python
ss_power
```

```python
ss_power_over_n = ss_power/156
```

```python
ss_power_over_n
```

## Getting the minimum value between min_hh_dd and ss_power_over_n 

```python
min_hh_dd.values
```

```python
ss_power_over_n
```

```python
power_allocated_cea_using_min_hh_dd = np.minimum(min_hh_dd.values, ss_power_over_n)
```
