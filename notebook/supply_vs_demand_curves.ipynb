```python
import numpy as np
import pandas as pd
```

```python
power_allocated_cea_using_min_hh_dd = pd.read_excel('power_allocated_cea_using_min_hh_dd.xlsx')
power_allocated_cea_using_max_hh_dd = pd.read_excel('power_allocated_cea_using_max_hh_dd.xlsx')
power_allocated_cea_using_avg_hh_dd = pd.read_excel('power_allocated_cea_using_avg_hh_dd.xlsx')
max_hh_dd = pd.read_excel('max_hh_demand_reformatted.xlsx')
min_hh_dd = pd.read_excel('min_hh_demand_reformatted.xlsx')
avg_hh_dd = pd.read_excel('avg_hh_demand_reformatted.xlsx')
demand_supply = pd.read_excel('Demand_Supply.xlsx')
```

```python
# sum columns for each table
sum_max_hh_dd = max_hh_dd.sum()
sum_min_hh_dd = min_hh_dd.sum()
sum_avg_hh_dd = avg_hh_dd.sum()
```

```python
# Power supplied
ss_power_values = demand_supply['SS_Power']
```

```python
# Actual power supplied
sum_power_allocated_cea_using_min_hh_dd = power_allocated_cea_using_min_hh_dd.sum()
sum_power_allocated_cea_using_max_hh_dd = power_allocated_cea_using_max_hh_dd.sum()
sum_power_allocated_cea_using_avg_hh_dd = power_allocated_cea_using_avg_hh_dd.sum()
```

## Min Curves

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

# Plot the mean with a very thick line
ax.plot(range(1, 25), ss_power_values, color='red', linewidth=1, label='Total Power Supplied')
ax.plot(range(1, 25), sum_power_allocated_cea_using_min_hh_dd, color='green', linewidth=1, label='Power Allocated')
ax.plot(range(1, 25), sum_min_hh_dd, color='blue', linewidth=1, label='Minimum Hourly Demand', linestyle="--")

# Shade the space between the two lines
ax.fill_between(range(1, 25), ss_power_values, sum_power_allocated_cea_using_min_hh_dd, color='lightgray', alpha=0.3)

# Set the axis labels and title
ax.set_xlabel('Hour')
ax.set_ylabel('Power Supply')
ax.set_title('Power Supply against Power Allocated & Minimum Required')

# legend
ax.legend()

# save
plt.savefig('power_supply_against_power_allocated_and_minimum_required.png')

# Show the plot
plt.show()
```

## Max Curves

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

# Plot the mean with a very thick line
ax.plot(range(1, 25), ss_power_values, color='red', linewidth=1, label='Total Power Supplied')
ax.plot(range(1, 25), sum_power_allocated_cea_using_max_hh_dd, color='green', linewidth=1, label='Power Allocated')
ax.plot(range(1, 25), sum_max_hh_dd, color='blue', linewidth=1, label='Maximum Hourly Demand', linestyle="--")

# Shade the space between the two lines
ax.fill_between(range(1, 25), ss_power_values, sum_power_allocated_cea_using_max_hh_dd, color='lightgray', alpha=0.3)

# Set the axis labels and title
ax.set_xlabel('Hour')
ax.set_ylabel('Power Supply')
ax.set_title('Power Supply against Power Allocated & Maximum Required')

# legend
ax.legend()

# save
plt.savefig('power_supply_against_power_allocated_and_maximum_required.png')

# Show the plot
plt.show()
```
